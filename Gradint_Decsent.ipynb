```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import make_regression

X,y = make_regression(n_samples=100, n_features=1, n_informative=1, n_targets=1,noise=20,random_state=15)
```

```python
plt.scatter(X,y)
print(X[1:10])
```

```python
class GDReggresser:
    def __init__(self,lr,epochs):
        self.m = 100
        self.b = 200
        self.lr = lr
        self.epochs = epochs
    def fit(self,X,y):
        for i in range(self.epochs):
            loss_m = -2 * np.sum((y - self.m * X.ravel() - self.b)*X.ravel())
            loss_b = -2 * np.sum(y - self.m*X.ravel() - self.b)
            self.m = self.m - (self.lr * loss_m)
            self.b = self.b - (self.lr * loss_b)
            print('epoch = ',i,'m = ',self.m,'b = ',self.b)
    def predict(self,X_test):
        return self.m*X_test + self.b

        
```

```python
GDR = GDReggresser(0.004,10)
GDR.fit(X,y)
```

```python
# m =  -2.851977382227096 b =  -3.0889604883684605
```

```python
from sklearn.metrics import r2_score
from sklearn.metrics import root_mean_squared_error
```

```python
X_result = GDR.predict(X)
r2_score(y,X_result)
```
