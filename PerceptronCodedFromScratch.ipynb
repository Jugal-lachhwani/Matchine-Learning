```python
from sklearn.datasets import make_classification
import numpy as np
import pandas as pd
```

```python
X, y = make_classification(n_samples=100, n_features=2, n_informative=1,n_redundant=0,
                           n_classes=2, n_clusters_per_class=1, random_state=41,hypercube=False,class_sep=10)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def step(x):
    if x >=0:
        return 1
    else:
        return 0
```

```python
def perceptron(X,y,epochs = 1000):
    weights = np.ones(X.shape[1]+1)
    lr = 0.1
    
    for _ in range(epochs):
        j = np.random.randint(0,X.shape[0])
        x = np.insert(X[j],0,1)
        y_hat = step(np.dot(x,weights))
        weights = weights + lr * (y[j] - y_hat)*x
    
    return weights
```

```python
j = np.random.randint(0,X.shape[0])
np.insert(X[j],0,1)
```

```python
perceptron(X,y)
intercept_ = perceptron(X,y)[0]
coef_ = perceptron(X,y)[[1,2]]
print(f"coef_ = {coef_}\nintercept = {intercept_}")
```

```python
m = -(coef_[0]/coef_[1])
b = -(intercept_/coef_[1])
```

```python
x_input = np.linspace(-3,3,100)
y_input = m*x_input + b
```

```python
plt.figure(figsize=(10,6))
plt.plot(x_input,y_input,color = 'red')
plt.scatter(X[:,0],X[:,1],c = y)
plt.ylim(-4,2.5)
```
